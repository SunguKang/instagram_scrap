```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Assuming df_combined2 has 'Name' and 'Media_count' columns

plt.figure(figsize=(15, 10))
ax = sns.barplot(data=df_combined.iloc[1:35], x='Name', y='Media_count', palette='deep')
plt.xlabel('Name')
plt.ylabel('post Count (Millions)')
plt.title('Total post number')

# Rotate x-axis labels
plt.xticks(rotation=45)

# Format y-axis ticks to display in millions
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:.0f}M'.format(x / 1000000)))
plt.savefig('data/post_count by streetwear.jpeg', format='jpeg', dpi=300)

plt.show()
```
